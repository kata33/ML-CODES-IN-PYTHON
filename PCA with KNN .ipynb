# Importing all  the Required Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

# Reading Csv File From Local Disk

```python
x=pd.read_csv("C:\\Users\\KATA\\Documents\\DATA SCIENCE\\DataSets 1\\trainpca.csv")
```

```python
x.head()
```

# Storing label Column in One of The Variable for Future Reference

```python
a=x['label']
```

# Dropping Label Column

```python
x.drop(["label"],axis=1,inplace=True)
```

# Performing Standard Scalar 

```python
s=StandardScaler()
```

```python
s1=s.fit_transform(x)
```

# Performing PCA

```python
p=PCA(0.95)
```

```python
p1=p.fit_transform(s1)
```

# Vizuvalizing PCA

```python
import matplotlib.pyplot as plt
plt.scatter(p1[:,0],p1[:,1],cmap="prism",c=a)
```

# Plotting The Elbow Curve 

```python
import numpy as np
plt.plot(np.cumsum(p.explained_variance_ratio_))
```

# Converting The Values Into a Data Frame

```python
v=pd.DataFrame(p1)
```

```python
v
```

```python
n=s.fit_transform(v)
```

```python
l=pd.DataFrame(n)
```

```python
l
```

# Concat The Label Column 

```python
f=pd.concat([a,l],axis=1)
```

```python
f
```

# Performing KNN on top of Pca since KNN works Well with High Dimension

```python
m=f.drop(["label"],axis=1)
n=f["label"]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=KNeighborsClassifier(n_neighbors=7)
```

# Splitting The Data Into Train And Test

```python
x_train,x_test,y_train,y_test=train_test_split(m,n,test_size=0.2,random_state=123)
```

```python
k.fit(x_train,y_train)
```

# Predecting The Multi Class Classification

```python
pred=k.predict(x_test)
```

# Evulating The Model

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```
